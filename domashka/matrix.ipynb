```python
class Matrix:
    def __init__(
        self,
        matrix = []
    ):
        self.matrix = matrix


    def __add__ (self: 'Matrix', other: list | float | int) -> 'Matrix':
        if isinstance(other, (Matrix, list)):
            if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
                ans_add = Matrix()
                ans_add = []
                for i in range (len(self.matrix)):
                    row_add = []
                    for k in range (len(self.matrix[i])):
                        el_add = self.matrix[i][k] + other.matrix[i][k]
                        row_add.append(el_add)
                    ans_add.append(row_add)
                return ans_add
        elif isinstance(other, (int, float)):
            ans_add = Matrix()
            ans_add = []
            for i in range (len(self.matrix)):
                row_add = []
                for k in range (len(self.matrix[i])):
                    el_add = self.matrix[i][k] + other
                    row_add.append(el_add)
                ans_add.append(row_add)
            return ans_add
        else:
            raise ValueError


    def __sub__ (self: 'Matrix', other: 'Matrix') -> 'Matrix':
        if isinstance(other, (Matrix, list)):
            if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
                ans_sub = Matrix()
                ans_sub = []
                for i in range (len(self.matrix)):
                    row_sub = []
                    for k in range (len(self.matrix[0])):
                        el_sub = self.matrix[i][k] - other.matrix[i][k]
                        row_sub.append(el_sub)
                    ans_sub.append(row_sub)
                return ans_sub
        elif isinstance(other, (int, float)):
            ans_sub = Matrix()
            ans_sub = []
            for i in range (len(self.matrix)):
                row_sub = []
                for k in range (len(self.matrix[i])):
                    el_sub = self.matrix[i][k] - other
                    row_sub.append(el_sub)
                ans_sub.append(row_sub)
            return ans_sub
        else:
            raise ValueError


    def transpose (self: 'Matrix') -> 'Matrix':
        ans_trans = []
        for i in range (len(self.matrix)):
            trans_row = []
            for k in range (len(self.matrix[0])):
                trans_row.append(self.matrix[k][i])
            ans_trans.append(trans_row)
        return ans_trans


    def T (self): 
        ans_trans = [] 
        for i in range(len(self.matrix)): 
            trans_row = [] 
            for k in range(len(self.matrix[0])): 
                trans_row.append(self.matrix[k][i]) 
            ans_trans.append(trans_row) 
        return Matrix(ans_trans)
        

    def __mul__ (self: 'Matrix', other: 'Matrix') -> 'Matrix':
        if isinstance(other, (Matrix, list)):
            if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
                ans_mul = Matrix()
                ans_mul = []
                for i in range (len(self.matrix)):
                    row_mul = []
                    for k in range (len(self.matrix[0])):
                        el_mul = self.matrix[i][k] * other.matrix[i][k]
                        row_mul.append(el_mul)
                    ans_mul.append(row_mul)
                return ans_mul
        elif isinstance(other, (int, float)):
            ans_mul = Matrix()
            ans_mul = []
            for i in range (len(self.matrix)):
                row_mul = []
                for k in range (len(self.matrix[i])):
                    el_mul = self.matrix[i][k] * other
                    row_mul.append(el_mul)
                ans_mul.append(row_mul)
            return ans_mul
        else:
            raise ValueError


    def __truediv__ (self: 'Matrix', other: list | float | int) -> 'Matrix':
            if isinstance(other, (Matrix, list)):
                if len(self.matrix) == len(other.matrix) and len(self.matrix[0]) == len(other.matrix[0]):
                    ans_div = Matrix()
                    ans_div = []
                    for i in range (len(self.matrix)):
                        row_div = []
                        for k in range (len(self.matrix[0])):
                            el_div = self.matrix[i][k] / other.matrix[i][k]
                            row_div.append(el_div)
                        ans_div.append(row_div)
                    return ans_div
            elif isinstance(other, (int, float)):
                ans_div = Matrix()
                ans_div = []
                for i in range (len(self.matrix)):
                    row_div = []
                    for k in range (len(self.matrix[0])):
                        el_div = self.matrix[i][k] / other
                        row_div.append(el_div)
                    ans_div.append(row_div)
                return ans_div
            else:
                raise ValueError


    def __matmul__(self: 'Matrix', other: 'Matrix') -> 'Matrix': 
        if len(self.matrix[0]) == len(other.matrix) and len(self.matrix) == len(other.matrix[0]):
            ans_matmul = [] 
            for i in range(len(self.matrix)): 
                row_matmul = [] 
                for j in range(len(other.matrix[0])): 
                    number = 0 
                    for k in range(len(other.matrix)): 
                        number += self.matrix[i][k] * other.matrix[k][j] 
                    row_matmul.append(number) 
                ans_matmul.append(row_matmul) 
            return ans_matmul
        else:
            raise ValueError
```

```python
%%ipytest

matrix1 = Matrix ([
    [1, 2, 3],
    [4, 5, 6],
    [7, 8, 9],
])

matrix2 = Matrix ([
    [9, 8, 7],
    [6, 5, 4],
    [3, 2, 1],
])

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])
def test_matrix_add_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (matrix1 + matrix2 == (np.array(matrix1.matrix) + np.array(matrix2.matrix))).all()
    ), 'add section with matrices is not working right'

@pytest.mark.parametrize("matrix1,number", [[matrix1, 13.2]])
def test_matrix_add_number(matrix1: Matrix, number: int | float):
    assert (
        (matrix1 + number == np.array(matrix1.matrix) + number).all()
    ), 'add section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])
def test_matrix_sub_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (matrix1 - matrix2 == np.array(matrix1.matrix) - np.array(matrix2.matrix)).all()
    ), 'sub section with matrices is not working right'

@pytest.mark.parametrize("matrix1,number", [[matrix1, 12.2]])    
def test_matrix_sub_number(matrix1: Matrix, number: int | float):
    assert (
        (matrix1 - number == np.array(matrix1.matrix) - number).all()
    ), 'sub section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1", [matrix1])    
def test_matrix_transpose(matrix1: Matrix):
    assert (
        (matrix1.transpose() == np.array(matrix1.matrix).T).all()
    ), 'transpose section is not working right'
    
# -----------------------------------------

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])
def test_matrix_mul_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (matrix1 * matrix2 == np.array(matrix1.matrix) * np.array(matrix2.matrix)).all()
    ), 'mul section with matrices is not working right'

@pytest.mark.parametrize("matrix1,number", [[matrix1, 13]])
def test_matrix_mul_number(matrix1: Matrix, number: int | float):
    assert (
        (matrix1 * number == np.array(matrix1.matrix) * number).all()
    ), 'mul section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2]])    
def test_matrix_div_matrix(matrix1: Matrix, matrix2: Matrix):
    assert (
        (matrix1 / matrix2 == np.array(matrix1.matrix) / np.array(matrix2.matrix)).all()
    ), 'div section with matrices is not working right'

@pytest.mark.parametrize("matrix1,number", [[matrix1, 2]])    
def test_matrix_div_number(matrix1: Matrix, number: int | float):
    assert (
        (matrix1 / number == np.array(matrix1.matrix) / number).all()
    ), 'div section with matrix and number is not working right'

@pytest.mark.parametrize("matrix1,matrix2", [[matrix1, matrix2.T()]])    
def test_matrix_matmul_matrix(matrix1: Matrix, matrix2: Matrix):
    assert(
        (matrix1 @ matrix2 == np.array(matrix1.matrix) @ np.array(matrix2.matrix)).all()
    ), 'matmul section is not working right, and don\'t forget about T() if needed'
```
